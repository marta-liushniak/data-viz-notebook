```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### QUALITATIVE DATA

```python
listings = pd.read_csv('listings.csv')
```

```python
listings
```

```python
# to create a bar chart that will display the count of Airbnb listings in each neighbourhood group of NY city 
sns.countplot(x = 'neighbourhood_group', data = listings)
plt.show()
```

```python
# to create a bar chart that will display the average price of Airbnb listings in each neighbourhood group of NY city 
sns.barplot(x = 'neighbourhood_group', y = 'price', data = listings)
plt.show()
```

```python
# to remove the confidence intervals from the previous bar chart
sns.barplot(x = 'neighbourhood_group', y = 'price', data = listings, errorbar = ('ci', False))
plt.show()
```

```python
plt.hist(listings['price'])
plt.xlabel('price (in US $)')
plt.show()
```
